## 4. Verificar los resultados en MLflow

````python
import mlflow

mlflow.set_tracking_uri(MLFLOW_TRACKING_URI)

# Obtener el ultimo run del experimento
experiment = mlflow.get_experiment_by_name(MLFLOW_EXPERIMENT_NAME)
runs = mlflow.search_runs(
    experiment_ids=[experiment.experiment_id],
    order_by=["start_time DESC"],
    max_results=5,
)

print(f"Ultimos runs del experimento '{MLFLOW_EXPERIMENT_NAME}':")
print("=" * 60)
for _, row in runs.iterrows():
    print(f"  Run ID  : {row['run_id'][:12]}...")
    print(f"  Accuracy: {row['metrics.accuracy']:.4f}")
    print(f"  Modelo  : {row['params.modelo']}")
    print(f"  Fecha   : {row['start_time']}")
    print()
````
